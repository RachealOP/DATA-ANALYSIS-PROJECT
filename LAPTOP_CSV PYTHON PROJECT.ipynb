## Brief Description of The Column
* num_cores: The number of tasks and programs a laptop can execute at once.
* num_threads: The number of a logical software unit that can run on a single core.
* ram_memory: means random access memory, its used as a short term memory storage for a computer. 
* gpu_brand: A graphics processing unit (GPU) is a printed circuit board responsible for processing image-related data.
* gpu_type: The type of GPU in this dataset are dedicated, integrated and apple, 
* display_size: also known as the viewable image size (VIS) is the physical size of the area where pictures and videos are displayed.
* OS: means (Operating System) is a system software that manages computer hardware and software resources.
* processor_tier: The processor_tier are categorized based on their performance, features and price range.

## Exploratory Data Analysis

## 1. Average price of laptop by brand

```python
# Calculate the average price by brand
average_price_by_brand = df.groupby('brand')['Price'].mean().reset_index().sort_values(by = "brand",ascending = False)

average_price_by_brand[0:10]
```

```python
# visualizing the average price of laptop by brand
plt.figure(figsize = (12,8))
sns.barplot(x = "brand", y = "Price", data =average_price_by_brand[0:10])
plt.title("Average Laptop Prices By Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.xticks(rotation = 45)
```

 Brands with higher average laptop prices are perceived as premium or luxury brands by consumers. Variations in average laptop prices across different brands may indicate segmentation within the market based on target demographics and consumer preferences. Some brands may target premium segments with high-end laptops, while others may focus on the mid-range or budget segments to appeal to a broader audience.

2. Average price of laptop

```python
# Calculate the average price of a laptop
average_price = df['Price'].mean()

print("Average Price of Laptops:", average_price)
```

## 3. The average price of laptop by rating

```python
# Group by rating and calculate the average price
average_price_by_rating = df.groupby('Rating')['Price'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(average_price_by_rating.index, average_price_by_rating.values, marker='o', linestyle='-')
plt.title('Average Price of Laptops by Rating')
plt.xlabel('Rating')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

## Variations in average prices by rating may reflect brand positioning strategies within the market. Higher average prices for laptops with higher ratings indicates that consumers are willing to pay more for laptops they perceive as having better quality, performance, and reliability. 


## 4. Average Prices of 10 Most Rated Laptop Brands

```python
# Find the 10 most rated brands
most_rated_brands = df['brand'].value_counts().nlargest(10)

# Calculate average price for each of the most rated brands
average_prices = []
for brand in most_rated_brands.index:
    average_price = df[df['brand'] == brand]['Price'].mean()
    average_prices.append(average_price)

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(most_rated_brands.index, average_prices)
plt.ylabel('Brand')
plt.xlabel('Average Price')
plt.title('Average Price for Top 10 Most Rated Brands')
plt.grid(axis='x')
plt.show()
```

## The chart illustrates the average prices of the top 10 most rated laptop brands, showcasing the differences in pricing strategies across brands. Some brands have position themselves as premium options with higher average prices, while others may target budget-conscious consumers with lower average prices.

## 5. Average Price across Different Processor_tier.

```python
# Calculate mean price for each processor tier
mean_prices = df.groupby('processor_tier')['Price'].mean().reset_index()
mean_prices
```

```python
# Create bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='processor_tier', y='Price', data=mean_prices, palette='viridis')
plt.title('Average Price Across Different Processor Tiers')
plt.xlabel('Processor Tier')
plt.ylabel('Mean Price (in $)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Different processor tiers typically represent varying levels of performance and capabilities. Comparing the average prices across different processor tiers allows for assessing the price-performance ratio of laptops. Consumers may evaluate the cost-effectiveness of laptops based on the performance benefits offered by different processor tiers. Brands that offer competitive prices relative to the performance provided by their processors may attract value-conscious consumers seeking optimal performance within their budget constraints.
